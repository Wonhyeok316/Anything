## COVID-19 백신 종류별 접종 인구

```python
date = [1, 2, 3] #1일 2일 3일
az = [2, 3, 4]
pfizer = [5, 1, 2]
moderna = [1, 8, 2]
```

```python
#혹은
plt.plot(date, az, label = 'az')
plt.plot(date, pfizer, label = 'pfizer', marker='o', linestyle='--')
plt.plot(date, moderna, label='moderna', marker='s', ls = '-.')

plt.legend(ncol=2)
```
