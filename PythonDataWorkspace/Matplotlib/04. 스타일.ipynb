# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=10)
```

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker = 'o',linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor='b', markerfacecolor='g')
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

## 색깔

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='g')
```

## 포맷

```python
plt.plot(x, y, 'ro--')# red, o, -- / color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go') #linestyle 없으면 선이 없어지네
```

```python
plt.plot(x, y, color='g', marker='o', linestyle='None')
```

## 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

```python
#축약어로 쓰이는 것들은 matplotlib markerfacecolor mfc 이렇게 구글에 찾아보자
```

## 투명도

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.7)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=50) #dotot per inch, 해상도
plt.plot(x, y)
```

## 배경색`

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x, y)
```
