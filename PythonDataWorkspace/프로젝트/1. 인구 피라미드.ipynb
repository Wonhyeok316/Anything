# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
df_m = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y ')
#위 3개 행은 필요없음
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int) #  ,를 공백으로
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel('201108_201108_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w
# .1이 뒤에 다 붙어있음. 앞에꺼랑 중복되서 그럼
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns=df_m.columns
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
df_w.iloc[0]
```

## 데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, df_m.iloc[0]//1000, alpha=0.3) # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0]//1000, alpha=0.3)
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0]//1000) # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.title('2011 대한민국 인구 피라미드')
plt.savefig('2011_인구피라미드.png', dpi=100)
plt.show()
```

```python
# 2021년 껏도 만들어 보자
import pandas as pd
df_m21 = pd.read_excel('202112_202112_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_w21 = pd.read_excel('202112_202112_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B, AB:AV')
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m21.columns, -df_m21.iloc[0]//1000)
plt.barh(df_m21.columns, df_w21.iloc[0]//1000)
plt.title('2021 대한민국 인구 피라미드')
plt.savefig('2021_인구피라미드.png', dpi=100)
plt.show()
```
