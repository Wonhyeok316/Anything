```python
# https://coding-kindergarten.tistory.com/115
import pandas as pd

df = pd.read_csv('premierLeague_tables_1992-2017.csv')
df.head()
```

```python
df = df[['season', 'team', 'points']]
df.head()
```

```python
df = df.pivot_table(values = 'points', index = 'season', columns = 'team')
#반드시 행이 시간이어야함.
df.head()
```

```python
#NAN값을 모두 0으로
df = df.fillna(0)
df
```

```python
for row in range(len(df.index)):
    if row>0:
        df.iloc[row, :] += df.iloc[row-1, :]
df.head()
```
